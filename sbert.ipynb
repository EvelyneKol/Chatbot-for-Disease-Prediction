```python
import pandas as pd

# symptoms of my dataset
data=pd.read_csv("Training.csv").drop("Unnamed: 133", axis=1)
list(data.columns.values)
list(data)
symptom_list = list(data)
print(symptom_list)
```

```python
from sentence_transformers import SentenceTransformer, util

# Initialize SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# αντικατάσταση underscore και lowercasing τα συμπτώματα
def preprocess_symptoms(symptoms):
    return [symptom.replace("_", " ").lower() for symptom in symptoms]

normalized_symptoms = preprocess_symptoms(symptom_list)

# eencode symptoms
symptom_embeddings = model.encode(normalized_symptoms)
```

```python
def map_symptoms(user_input):
    # lowercase user input
    processed_input = user_input.lower()

    # encode input
    input_embedding = model.encode([processed_input])

    # ypologismos similarities
    similarities = util.pytorch_cos_sim(input_embedding, symptom_embeddings)
    matches = [
        (normalized_symptoms[i], similarities[0][i].item())
        for i in range(len(normalized_symptoms))
    ]

    # matching based on the threshold
    threshold = 0.6
    matches = [symptom for symptom, score in matches if score > threshold]

    return matches

# example user input
user_input = "I have difficulty breathing and pain in my chest"
matched_symptoms = map_symptoms(user_input)
print(matched_symptoms)
```

```python
from sentence_transformers import SentenceTransformer, util
import tensorflow_hub as hub
import numpy as np

# Example models
models = {
    "SBERT-MiniLM": SentenceTransformer('all-MiniLM-L6-v2'),
    "SBERT-MPNet": SentenceTransformer('all-MPNet-base-v2'),
    "SBERT-multi-qa-mpnet": SentenceTransformer('multi-qa-mpnet-base-dot-v1'),
    "USE": hub.load("https://tfhub.dev/google/universal-sentence-encoder/4"),
}

# Preprocess symptoms
def preprocess_symptoms(symptoms):
    return [symptom.replace("_", " ").lower() for symptom in symptoms]

# Symptom list
normalized_symptoms = preprocess_symptoms(symptom_list)

# Encode symptoms using all models
embeddings = {}
for model_name, model in models.items():
    if model_name == "USE":
        embeddings[model_name] = np.array(model(normalized_symptoms))
    else:
        embeddings[model_name] = model.encode(normalized_symptoms)

# Similarity computation
def map_symptoms(user_input, model_name):
    processed_input = user_input.lower()
    if model_name == "USE":
        input_embedding = np.array(models[model_name]([processed_input]))
    else:
        input_embedding = models[model_name].encode([processed_input])

    similarities = util.pytorch_cos_sim(input_embedding, embeddings[model_name])
    matches = [
        (normalized_symptoms[i], similarities[0][i].item())
        for i in range(len(normalized_symptoms))
    ]
    threshold = 0.6
    return [symptom for symptom, score in matches if score > threshold]

# User input
user_input = "I feel chest discomfort and have trouble breathing"

# Compare results
results = {}
for model_name in models:
    results[model_name] = map_symptoms(user_input, model_name)

# Print results
for model_name, matched_symptoms in results.items():
    print(f"Model: {model_name}")
    print(f"Matched Symptoms: {matched_symptoms}")
    print()
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# List of different SBERT models to compare
sbert_models = [
    'all-mpnet-base-v2',         # General-purpose model
    'multi-qa-mpnet-base-dot-v1',  # Multilingual model
    'all-distilroberta-v1',  # Optimized for paraphrase similarity
    'all-MiniLM-L12-v2',
    'multi-qa-distilbert-cos-v1',  # Larger model optimized for paraphrases,
    'all-MiniLM-L6-v2',
]

# Function to preprocess symptoms: replace underscores and lowercase
def preprocess_symptoms(symptoms):
    return [symptom.replace("_", " ").lower() for symptom in symptoms]

# Function to map user input to symptoms using different SBERT models
def map_symptoms(user_input, model_name):
    # Initialize the model
    model = SentenceTransformer(model_name)

    # Preprocess user input
    processed_input = user_input.lower()

    # Encode the user input
    input_embedding = model.encode([processed_input])

    # Preprocess and encode symptoms
    normalized_symptoms = preprocess_symptoms(symptom_list)
    symptom_embeddings = model.encode(normalized_symptoms)

    # Calculate cosine similarities
    similarities = util.pytorch_cos_sim(input_embedding, symptom_embeddings)

    # Get matches based on a threshold
    threshold = 0.50
    matches = [
        (normalized_symptoms[i], similarities[0][i].item())
        for i in range(len(normalized_symptoms))
    ]
    matches = [symptom for symptom, score in matches if score > threshold]

    return matches

# Example user input
user_input = "I feel pain all over my body and in my bones"

# Compare different SBERT models
for model_name in sbert_models:
    print(f"Results for model: {model_name}")
    matched_symptoms = map_symptoms(user_input, model_name)
    print("Matched Symptoms:", matched_symptoms)
    print("-" * 50)
```
