```python
import pandas as pd

# Διάβασμα CSVs
df1 = pd.read_csv('dataset1.csv')  # CSV No1
df2 = pd.read_csv('Training.csv')  # CSV No2

# Ένωση των δύο DataFrames
merged_df = pd.concat([df1, df2], axis=0, ignore_index=True).fillna('')

# Ένωση των ασθενειών στη στήλη "Disease"
merged_df['Disease'] = merged_df['Disease'] + ' ' + merged_df['prognosis']

# Drop των στηλών 'prognosis' και 'Unnamed: 133'
merged_df = merged_df.drop(columns=['prognosis', 'Unnamed: 133'])

# Αντικατάσταση μη αριθμητικών τιμών με 0 και μετατροπή σε ακέραιους
one_hot_columns = merged_df.columns.difference(['Disease'])  # όλες οι στήλες εκτός από τη "Disease"
merged_df[one_hot_columns] = merged_df[one_hot_columns].replace('', 0).astype(int)

merged_df.head()
```

```python
# αποθήκευση των ενοποιημένων δεδομένων σε csv 
merged_df.to_csv('Dataset.csv', index=False)
```

```python
#Μετατροπή όλων των στηλών σε lowercase
# Ελέγχουμε κάθε στήλη αν είναι τύπου string και την μετατρέπουμε
merged_df = merged_df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

# Αποθήκευση του DataFrame σε νέο CSV αν χρειάζεται
merged_df.to_csv('Dataset.csv', index=False)
```
