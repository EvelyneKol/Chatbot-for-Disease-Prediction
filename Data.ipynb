```python
import numpy as np
import pandas as pd
```

```python
#διαβασμα δεδομενωνν
data = pd.read_csv('dataset.csv')
print(data.head())
```

```python
# χωρίζουμε ασθένειες/συμπτώματα με ίδιο όνομα που έχουν '^' σε κεχωριστές γραμμές
data['Symptom'] = data['Symptom'].str.split('^')
data = data.explode('Symptom').reset_index(drop=True)
data['Disease'] = data['Disease'].str.split('^')
data = data.explode('Disease').reset_index(drop=True)

print(data.head())
```

```python
# αποθηκευουμε τα μοναδικα ονόματα ασθενειών εγγραφών για το One hot encoding
unique_diseases = data['Disease'].unique()
unique_symptoms = data['Symptom'].unique()
```

```python
# δημιορυγία df για το One hot encoding
one_hot_df = pd.DataFrame(0, index=unique_diseases, columns=unique_symptoms)
```

```python
# για κάθε γραμμή των δεδομένων (data df) βάζουμε 1 αν το σύμπτωμα υπάρχει στην τρέχουσα ασθένεια 
for _, row in data.iterrows():
    one_hot_df.at[row['Disease'], row['Symptom']] = 1
```

```python
x = one_hot_df[cols]
y = one_hot_df.iloc[:, 0]  
```

```python
# split σε training και testing 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
# RANDOM FOREST 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Αρχικοποίηση του Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42) 

# Εκπαίδευση 
rf.fit(x_train, y_train)

# Πρόβλεψη 
y_pred = rf.predict(x_test)

# ακρίβεια μοντέλου
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Αρχικοποίηση του k-NN Classifier
knn = KNeighborsClassifier(n_neighbors=5)  # n_neighbors: αριθμός των γειτόνων (προεπιλογή 5)

# Εκπαίδευση 
knn.fit(x_train, y_train)

# Πρόβλεψη 
y_pred = knn.predict(x_test)

# Αξιολόγηση με ακρίβεια
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# MULTINOMIAL NAIVE BAYES 
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Αρχικοποίηση του Multinomial Naive Bayes
nb = MultinomialNB()

# Εκπαίδευση του μοντέλου 
nb.fit(x_train, y_train)

# Πρόβλεψη αποτελεσμάτων 
y_pred = nb.predict(x_test)

# Αξιολόγηση του μοντέλου 
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# SVM
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Αρχικοποίηση του SVM classifier
svm = SVC(kernel='linear')  

# Εκπαίδευση μοντέλου
svm.fit(x_train, y_train)

# Πρόβλεψη
y_pred = svm.predict(x_test)

# Αξιολόγηση
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (SVM): {accuracy * 100:.2f}%")
```

```python
# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Αρχικοποίηση του Logistic Regression classifier
logreg = LogisticRegression()

# Εκπαίδευση μοντέλου
logreg.fit(x_train, y_train)

# Πρόβλεψη
y_pred = logreg.predict(x_test)

# Αξιολόγηση
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (Logistic Regression): {accuracy * 100:.2f}%")
```

```python
# DECISION TREE
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Αρχικοποίηση του Decision Tree classifier
dt = DecisionTreeClassifier()

# Εκπαίδευση μοντέλου
dt.fit(x_train, y_train)

# Πρόβλεψη
y_pred = dt.predict(x_test)

# Αξιολόγηση
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (Decision Tree): {accuracy * 100:.2f}%")
```
